```python
import pandas as pd

features = pd.read_csv("../data/raw/simple_data.csv")
features = features[features.Year==2019]

new_feats = pd.DataFrame()
new_feats['Country or Area Code'] = features['Country or Area Code'].drop_duplicates()
new_feats = new_feats.set_index('Country or Area Code')

# create new columns in new_feats for each area type and populate with ele_proportion
for value in features['Area_type'].drop_duplicates().values:
    new_feats = pd.concat([
        new_feats, 
        features[
            features[
                'Area_type']==f"{value}"][['Elec_proportion', 'Country or Area Code']].set_index(
                        'Country or Area Code').rename(columns={'Elec_proportion': f"{value}"})],
        axis=1
        )
# drop the countries which have na data
new_feats.dropna(axis=0, inplace=True)


labels = pd.read_csv("../data/labels.csv")
labels = labels[:-1]
labels['Country or Area Code'] = pd.to_numeric(labels['Country or Area Code'])
labels.set_index('Country or Area Code', inplace=True)

data = new_feats.join(
    labels, how='left')

data
# data.dropna(axis=0, inplace=True)
```
